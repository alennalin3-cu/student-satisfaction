## Ranking vs Tuition Measures

To test the robustness of this relationship, both in-state and out-of-state tuition are examined separately.

### Do schools with higher in-state tuition tend to rank better?

```python
import matplotlib.pyplot as plt

df["Tuition (In-State)"] = pd.to_numeric(df["Tuition (In-State)"])

df.plot(
    x = "Tuition (In-State)",
    y = "Public Ranking",
    kind = "scatter"
)

plt.gca().invert_yaxis
plt.title("In-State Tuition vs US News Public Ranking")
plt.show()
```

### Do schools with higher out-of-state tuition tend to rank better?

```python
import matplotlib.pyplot as plt

df["Tuition (Out-of-State)"] = pd.to_numeric(df["Tuition (Out-of-State)"])

df.plot(
    x = "Tuition (Out-of-State)",
    y = "Public Ranking",
    kind = "scatter"
)

plt.gca().invert_yaxis
plt.title("Out-of-State Tuition vs US News Public Ranking")
plt.show()
```

There is no relationship based on this visualization. However, something interesting observed is that the two most expensive out-of-state public schools (bottom right) rank within the top 5. Similarly, the two cheapest out-of-state public schools (top left) rank near the worst out of the 50 schools analyzed. It can also be seen that most of the top-10 schools charge more than the mean amount for out-of-state tuition (~$40k).

```python
df[
    ["Public Ranking", "Tuition (In-State)", "Tuition (Out-of-State)"]
].corr()
```

### Conclusion:
Across both tuition measures, correlations with ranking are weak, indicating no strong linear relationship between institutional ranking and tuition levels.

### Interpretation:
While tuition is often treated as an explanatory value for ranking, this analysis considers ranking as a potential driver for tuition prices, reflecting the idea that a better (lower) ranking enables schools to charge more. However, the absence of strong data suggests that out-of-state tuition pricing for public schools is more constrained by factors other than ranking alone.

### Potential Reasons for the Lack of Relationship
- State governments often regulate tuition increases
- Political or accesibility pressures
- Differences in regional cost-of-living
- Some institutions rely more heavily on government funding, while others rely more on tuition revenue alone

As a result, tuition does not appear to function as a market-driven price signal in the same way it might for private institutions.

## Do schools with a higher percentage of Pell students have less first-year students living on campus?

```python
correlation = df['Awarded Pell (22-23)'].corr(df['% First Years Living On-Campus']).round(5)
print(f'The correlation between Pell students and first-years living on campus is {correlation}.')
```

### Conclusion
There is a strong relationship between the percentage of Pell students and the percentage of first-year students living on campus. The more Pell students there are, the less likely it is for first-years to live on campus.
### Interpretation
The majority of public schools do not require first-years to live on campus. Students who receive Pell grants come from lower-income backgrounds, meaning they have less financial freedom regarding their college decisions, including housing, which is oftentimes more expensive than tuition. To cut costs, students on Pell may choose to commute from home if possible, and if a school requires dorming, they may choose not attend at all. Thus, if a school is more welcoming to Pell recipients, they will have to make a trade-off with decreases in first-year housing.

## Is 25th percentile, average, or 75th percentile GPA of admitted students more closely related to public ranking?

```python
df['25th Percentile GPA'] = pd.to_numeric(df['25th Percentile GPA'], errors="coerce")
df['Average High School GPA'] = pd.to_numeric(df['Average High School GPA'], errors="coerce")
df['75th Percentile GPA'] = pd.to_numeric(df['75th Percentile GPA'], errors="coerce")

correlation_25 = df['25th Percentile GPA'].corr(df['Public Ranking'])
correlation_avg_gpa = df['Average High School GPA'].corr(df['Public Ranking'])
correlation_75 = df['75th Percentile GPA'].corr(df['Public Ranking'])

print(f'The correlation between Public Ranking and 25th Percentile GPA is {correlation_25.round(5)}.')
print(f'The correlation between Public Ranking and Average GPA is {correlation_avg_gpa.round(5)}.')
print(f'The correlation between Public Ranking and 75th Percentile GPA is {correlation_75.round(5)}.')
```

### Conclusion
25th Percentile GPA had the strongest correlation with public ranking (-0.6238), followed by 75th Percentile GPA (-0.42314), with average GPA showing the weakest relationship (-0.38165).

### Interpretation
When looking at the raw data, it can be seen that many universities, regardless of their ranking, have similarly high average and 75th percentile GPAs. However, 25th percentile GPA has a clearer trend of dropping as ranking drops. The stronger relationship between ranking and the 25th percentile GPA suggests that selectivity at the lower bound of admitted students may be more closely tied to institutional ranking than the academic profile of top admits. Higher-ranked institutions appear to maintain higher minimum academic thresholds, even if their top admits are similar across schools. While these correlations do not imply causation, they indicate that rankings are more closely aligned with the academic baseline of admitted students than with peak performance.



```python
df.plot(
    x="Public Ranking",
    y="25th Percentile GPA",
    kind="scatter"
)

plt.gca().invert_xaxis()
plt.title("Public Ranking vs 25th Percentile Admitted GPA")
plt.xlabel("Public Ranking (Lower = Better)")
plt.ylabel("25th Percentile GPA")
plt.show()
```

## Do AAU schools rank more highly?

### Hypotheses:
- **Null Hypothesis (Hâ‚€):** There is no difference in the distribution of public university rankings between AAU and non-AAU schools.
- **Alternative Hypothesis (Hâ‚):** AAU schools tend to have better (lower) public rankings than non-AAU schools.

Because public rankings are ordinal and not normally distributed, a Mannâ€“Whitney U test was used to compare the distributions of rankings between AAU and non-AAU schools.


```python
from scipy.stats import mannwhitneyu

# Encode AAU membership
df['AAU Member?'] = df['AAU Member?'].map({'Yes': 1, 'No': 0})
df = df.dropna(subset=['Public Ranking', 'AAU Member?'])

aau_ranks = df.loc[df['AAU Member?'] == 1, 'Public Ranking']
non_aau_ranks = df.loc[df['AAU Member?'] == 0, 'Public Ranking']

u_stat, p_value = mannwhitneyu(
    aau_ranks,
    non_aau_ranks,
    alternative="less"
)

print("U statistic:", u_stat)
print("p-value:", round(p_value, 8))
print("\nAAU member median ranking:", aau_ranks.median())
print("Non-AAU member median ranking:", non_aau_ranks.median())

if p_value < 0.05:
    print("\nWe reject the null hypothesis. The difference is statistically significant.")
else:
    print("\nWe fail to reject the null hypothesis.")
```

### Conclusion
The results of the Mannâ€“Whitney U test suggest that AAU institutions tend to have better public rankings than non-AAU institutions. This finding is consistent with the broader role of AAU membership as a marker of research intensity and institutional prestige, though causality cannot be inferred.
